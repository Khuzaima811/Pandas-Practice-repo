### Data Wrangling :Join,Combine,and Reshape

#### 8.2 Combining and Merging DataSets

```python
d0=pd.read_excel("test_data.xlsx",sheet_name=0)
d1=pd.read_excel("test_data.xlsx",sheet_name=1)
d2=pd.read_excel("test_data.xlsx",sheet_name=2)
```

```python
d01=pd.concat([d0,d1],axis=0)
```

```python
d01.reset_index(drop=True,inplace=True)
```

```python
d01
```

```python
d02=pd.concat([d0,d2],axis=1)
```

```python
d02
```

```python
pd.merge(d0,d2,on="sales_man_id").drop("sales_man_name_y",axis=1)
```

```python
pd.merge(d0,d2,on="sales_man_name")     #The default merge type is inner
```

## Outer

```python
pd.merge(d0,d2,on="sales_man_id",how="outer")
```
