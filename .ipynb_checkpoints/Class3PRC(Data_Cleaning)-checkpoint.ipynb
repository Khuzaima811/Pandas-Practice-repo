### Handling Missing Values
- Drop all records having missing values
- Drop all columns having missing data
- Fill missing - hard coded value and Calculate statistically mean,mediun,mode
  
- bfill ,ffill

```python
data.dropna() #Dropping all row having NAN value
```

```python
data.dropna(axis=1)
```

```python
data.dropna(how="any",axis=1)
data.dropna(how="all",axis=1)
data.dropna(subset=["Department","Salary"],how="all",axis=0) #for that row ,having department and Salary both are NAN
```

### Filling Missing Data

```python
data.Department
```

```python
data.Department.mode()[0]
```

```python
data.Department.fillna(data.Department.mode()[0])
```

```python
data.Age.mean()
```

```python
data.Age.median()
```

```python
data.Age.fillna(data.Age.median())
```

```python
data.Salary.fillna(data.Salary.median())
```

```python
data.fillna({"Salary":data.Salary.median(),
            "Age":data.Age.median(),
            "Department":data.Department.mode()[0],
            "Designation":data.Designation.mode()[0]},inplace=True)
```

```python
def detect_outlier(data):
    Q1=np.percentile(data, 25)
    Q3=np.percentile(data, 75)
    IQR=Q3-Q1
    lb=Q1-1.5*IQR
    ub=Q3+1.5*IQR
    return lb,ub
```

```python
detect_outlier(data.Salary)[0]
```

```python
detect_outlier(data.Salary)[1]
```

```python
data[(data.Salary<=detect_outlier(data.Salary)[0]) | (data.Salary>=detect_outlier(data.Salary)[1])]
```

```python
data[(data.Age<=detect_outlier(data.Age)[0]) | (data.Age>=detect_outlier(data.Age)[1])]
```

```python
detect_outlier(data.Salary)
```
