```python
emp = pd.DataFrame({'emp_name':['nasir','asad','fahan' ,'noman', 'ali', 'ahmer','ahmed', 'hasan', 'umer', 'jawed', 'fahad' , 'rehan', 'tahir'],
                'country':['pak', 'pak', 'ind', 'usa', 'iran', 'iraq', 'bangla', 'uae', 'uae', 'iraq','iran','usa','pak']})
```

```python
country_capital = {'pak':'islambad', 'ind':'dehli','usa':'washington','iran':'tehran', 'iraq':'baghdad', 'uae':'abudhabi','bangla':'dhaka'}
```

```python
emp["capital"]=emp.country.map(country_capital)
```
