## Data Frame
  - 2 Dimensional Data
  - Data with more than one column

```python
result =     {"Roll":[12,32,43,54,54,75,32,56,67,34],
             "Name":["Khuzaima","Aman","Yasir","Tahir","Kamran","Taha","Talha","Saeed","Zubair","Irfan"],
             "Python":[23,34,35,54,76,98,23,56,78,89],
             "Pandas":[23,43,54,31,85,95,34,56,89,65],
             "Statistics":[23,24,57,78,96,54,90,87,56,34],
             "MachineLearning":[21,75,96,54,34,23,45,89,65,43],
             "DeepLearning":[21,85,65,34,12,87,89,76,56,43]} 
result
```

```python
result=pd.DataFrame(result)
result
```

## Working With Columns

```python
result['Total']=result.Python + result.Pandas + result.Statistics + result.MachineLearning + result.DeepLearning
result
```

```python
result
```

```python
result["Percentage"]=result.Total/500*100
result
```

##### By Using List Comprehension

```python
result["Grade"]= ["A+" if per>=70 else "A" if per>=60 else "B" if per>=50 else "C" if per>=40 else "D" if per>=30 else "E" if per>=20 else "F"  for per in result.Percentage]
```

```python
result
```

#### By Lamda Function
- They are Anonymous Function/Inline function

```python
result["Grade1"]=result.Percentage.apply(lambda per:"A+" if per>=70 else "A" if per>=60 else "B" if per>=50 else "C" if per>=40 else "D" if per >=30 else "E" if per>=20 else "F")
```

```python
result
```

### Outline Function

```python
def calculate_grade(per):
    if per>=70:
        return "A+"
    elif per>=60:
        return "A"
    elif per>=50:
        return "B"
    elif per>=40:
        return "C"
    elif per>=30:
        return "D"
    elif per>=20:
        return "E"
    else:
        return "F"
```

```python
result["Grade2"]=result.Percentage.apply(calculate_grade)
```

```python
result
```

```python
del result["Grade1"]
del result["Grade2"]
```

```python
result
```
