```python
def decide_tax_percentage(sal):
    if sal>=200000:
        return 20
    if sal>=150000:
        return 15
    if sal>=100000:
        return 10
    if sal>=50000:
        return 5
    else:
        return 0
```

```python
emp1["tax%"]=emp1.Salary.apply(decide_tax_percentage)
```

```python
emp1["tax%"]= emp1.Salary.apply(lambda x: 20 if x>=200000 else 15 if x>=150000 else 10 if x>=100000 else 5 if x>=50000 else 0) #There x refers the variable that stores each values like sal in outline function
```

```python
def decide_tax_amt(sal):
    if sal>=200000:
        return sal*.20
    elif sal>=150000:
        return sal*.15
    elif sal>=100000:
        return sal*.10
    elif sal>=50000:
        return sal*.05
    else:
        return 0
```

```python
emp1["Tax_Amount"] = emp1.Salary.apply(decide_tax_amt)
```

```python
emp1["Tax_Amount"]=emp1.Salary.apply(lambda x: x*.20 if x>=200000 else x*.15 if x>=150000 else x*.10 if x>=100000 else x*.05 if x>=50000 else 0)
```

```python
emp1
```

```python
emp1["Net"]=emp1.Salary - emp1.Tax_Amount
```

```python
result["Grade1"]=result.Percentage.apply(lambda per:"A+" if per>=70 else "A" if per>=60 else "B" if per>=50 else "C" if per>=40 else "D" if per >=30 else "E" if per>=20 else "F")
```

### Outline Function

```python
def calculate_grade(per):
    if per>=70:
        return "A+"
    elif per>=60:
        return "A"
    elif per>=50:
        return "B"
    elif per>=40:
        return "C"
    elif per>=30:
        return "D"
    elif per>=20:
        return "E"
    else:
        return "F"
```

```python
result["Grade2"]=result.Percentage.apply(calculate_grade)
```

```python
result
```
